```python
df_cancer=pd.DataFrame(np.c_[cancer["data"],cancer["target"]],columns=np.append(cancer["feature_names"],["target"]))
df_cancer
#data set after forming it into dataframe
```

```python
sns.pairplot(df_cancer,hue="target",vars=['mean smoothness','mean perimeter','mean texture','mean area','mean radius'])
#pairplots between 4 features mean radius,mean texture,mean perimeter , mean area
```

```python
sns.countplot(df_cancer["target"])
#histogram representing number of 0's and 1's in target
```

```python
plt.figure(figsize=(30,10))
sns.heatmap(df_cancer.corr(),annot=True)
#heat map representing corelations between each feature
```

```python
x=df_cancer.drop(["target"],axis=1)
x
```

```python
y=df_cancer["target"]
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=5)
#spliting data intlo test and train data
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
svc_model=SVC()
```

```python
svc_model.fit(X_train,y_train)
#Training model without preprocessing the data using Support vector classfier
```

```python
y_predict=svc_model.predict(X_test)
```

```python
y_predict
#predicted output by svm model before optimization the data
```

```python
cm=confusion_matrix(y_test,y_predict)
```

```python
sns.heatmap(cm,annot=True)
            #actual
            #predict
            #SO TP,FP
            #FN,TN
            #confusion matrix
```

```python
print(classification_report(y_test,y_predict))
#classification report before optimizing the data
```

# DATA PRE-PROCESSING

# optimazation -1- Normalisation

```python
min_train=X_train.min()
```

```python
range_train=(X_train-min_train).max()
```

```python
X_train_scaled=(X_train-min_train)/range_train
```

```python
sns.scatterplot(x=X_train["mean area"],y=X_train["mean smoothness"],hue=y_train)
#scatter plot before normalisation
```

```python
sns.scatterplot(x=X_train_scaled["mean area"],y=X_train_scaled["mean smoothness"],hue=y_train)
#feature scaling or unity normalisation
#scatter plot after unity normalisation , we can see x-axis values changed thousands to decimal range of 0-1
```

```python
min_test=X_test.min()
```

```python
range_test=(X_test-min_test).max()
```

```python
X_test_scaled=(X_test-min_test)/range_test
```

```python
svc_model.fit(X_train_scaled,y_train)
y_predict=svc_model.predict(X_test_scaled)
y_predict
#predicted output by svm model after normalising the data
```

```python
cm=confusion_matrix(y_test,y_predict)
sns.heatmap(cm,annot=True)
#confusion matrix
```

```python
print(classification_report(y_test,y_predict))
#classification report after optimizing the data 1 time 
```

# optimization-2-SVM gamma parameter

```python
param_grid={"C":[0.1,1,10,100],"gamma":[1,0.1,0.01,0.001],"kernel":["rbf"]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid=GridSearchCV(SVC(),param_grid,refit=True,verbose=4)
```

```python
grid.fit(X_train_scaled,y_train)
```

```python
grid.best_params_
```

```python
grid_predictions=grid.predict(X_test_scaled)
grid_predictions
#predicted output by svm model after normalising the data and using svm gamma parameter
```

```python
cm= confusion_matrix(y_test,grid_predictions)
sns.heatmap(cm,annot=True)
#confusion matrix after normalising the data and using svm gamma parameter
```

# Final Output 

```python
print(classification_report(y_test,grid_predictions))
#classification report after optimizing the data 2 times 
```
